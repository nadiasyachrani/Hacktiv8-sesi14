## Principal Component Analysis (CPA)
* misal punya 3 variabel untuk melihat hubungan antar variabelnya =>> 1 ruang 3D
* konsep : mengurangi dimensi ruang variabel yang ada pada data kita => berubah jadi 2 variabel

keuntungan/advantage
* bisa visualisasi dari data yang banyak
* mengurangi dampak kolineariti dari model


kekurangan/disadvantage
* ada outliers =>> pca bisa terdampak
* harus di scaling


```python
# import library
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
breast = load_breast_cancer()
```

```python
breast.feature_names
```

```python
breast.target_names #malignant:berbahaya, benign:tidak
```

```python
breast_labels = breast.target
```

```python
df = pd.DataFrame(breast.data,columns=breast.feature_names)
df
```

```python
df['target'] = breast.target
df
```

```python
# label asli format 0,1 mengubah label menjadi benign dan malignant
df['target'].replace(0, 'Benign', inplace=True)
df['target'].replace(1, 'Malignant', inplace=True)
```

```python
df.target.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
x = df.drop('target',axis=1)
stdscale = StandardScaler()
x_scaled = stdscale.fit_transform(x)
x_scaled
```

```python
df_feature_scaled = pd.DataFrame(x_scaled,columns=breast.feature_names)
df_feature_scaled
```

## PCA Sklearn
merangkup 30 kolum jadi 2 kolom

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
features_pca = pca.fit_transform(x_scaled)
features_pca
```

```python
pca_df = pd.DataFrame(features_pca,columns=['PC1','PC2'])
pca_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(20,20))
sns.heatmap(df_feature_scaled.corr(),annot=True)
```

0,957 udah multikoliner =>> sangat2 cocok untuk PCA

data banyak multikolinear

```python
# lihat variation ratio
print('Explain Variation per Principal Component', pca.explained_variance_ratio_)
```

pada pc1 menampung 44,2% dari informasi data, sedangkan pc2 menampung 18,9% dari data, sehingga ada 36%(sisanya) information loss(kkontribusinya tidak signifikan/tidak terlalu penting)

### Visualisasi data setelah PCA

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
targets=['Benign','Malignant']
colors = ['r','g']
for target,color in zip(targets,colors):
    indicesToKeep = df['target'] == target
    plt.scatter(pca_df.loc[indicesToKeep,'PC1'],
               pca_df.loc[indicesToKeep,'PC2'],
               c=color,s=40)
plt.legend(targets)
plt.title('PCA of Breast Cancer Data')
plt.xlabel('PC1')
plt.ylabel('PC2')

plt.show()
```

pc1 dan pc2 bisa mengimprove performance dari data

## K-Mean Clustering


K-means clustering adalah salah satu algoritma unsupervised learning yang populer digunakan untuk mengelompokkan data ke dalam kelompok-kelompok atau cluster berdasarkan kesamaan fitur atau karakteristik.

* gaada prediksi yang dihasilkan
* contoh : klasifikasi dari customer
* knn : dari dat ayg mau diprediksi, cari tetangga yang paling deket dari data tersebut, kelas mana yang paling banyak
* k-means : sebaran data, mennetukan k nya berapa/clusternya berapa, yang paling banyak jumlah datanya
* k nya ada 2, memilih titik2 random, memilihnya secara acak
* ketika udah pilih, dia akan tarik titik terdekat dari sentroit
* pasti ada iterasi, pake nilai tengah, rata-rata





### K-Means with SKlearn

```python
from sklearn.cluster import KMeans
```

```python
X = -2 * np.random.rand(100,2)
X1 = 1 + 2 * np.random.rand(50,2)
X[50:100,:] = X1

plt.scatter(X[:,0],X[:,1],s=40,c='b')
plt.show()
```

```python
pip install -U threadpoolctl
```

```python
inertia = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',
                   max_iter=300,n_init=10,random_state=0)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
    
plt.plot(range(1,11),inertia)
plt.title('Elbow Method')
plt.xlabel('No. of Clusters')
plt.ylabel('Inertia')
plt.show()
```

nilai optimum nya ada di 2

elbow method = cluster yang paling bagus itu dimana

inertia terkecil dan cluster terkecil

```python
kmeans_best = KMeans(n_clusters=2)
kmeans_best.fit(X)
```

```python
kmeans_best.cluster_centers_
```

```python
plt.scatter(X[:,0],X[:,1],s=40,c='b')
plt.scatter(1.95842126,1.99892219,s=40,c='r')
plt.scatter(-0.89850495,-0.97444104,s=40,c='g')
```

### KMeans dengan real data

```python
df_wholesale = pd.read_csv('Wholesale customers data.csv')
```

```python
df_wholesale
```

```python
df_wholesale.info()
```

```python
df_wholesale.describe()
```

```python
# Standardzing the data
from sklearn.preprocessing import StandardScaler

stdcale = StandardScaler()
wholesale_scaled = stdscale.fit_transform(df_wholesale)

pd.DataFrame(wholesale_scaled,columns=df_wholesale.columns).describe()
```

* proporsi sama
* standard deviasinya sama 1
* mean nya semuanya mendekati 0

```python
kmeans_2 = KMeans(n_clusters=2,init='k-means++')

kmeans_2.fit(wholesale_scaled)
```

```python
kmeans_2.inertia_
```

cukup tinggi

```python
inertia = []
for k in range(1,25):
    kmeans = KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(wholesale_scaled)
    inertia.append(kmeans.inertia_)
    
plt.plot(range(1,25),inertia)
plt.title('Elbow Method')
plt.xlabel('Jumlah Clusters')
plt.ylabel('Inertia')
plt.show()
```

misal ambil 5-8 =>> tengahnya 6, inertianya 1000 sekian

```python
kmeans_6 = KMeans(n_clusters=6,init='k-means++')
kmeans_6.fit(wholesale_scaled)
pred = kmeans_6.predict(wholesale_scaled)
```

```python
pred
```

```python
df_scaled = pd.DataFrame(wholesale_scaled,columns=df_wholesale.columns)
df_scaled
```

```python
df_scaled['clusters'] = pred
df_scaled
```

```python
df_scaled.clusters.value_counts()
```

```python
df_wholesale['clusters'] = pred
df_wholesale
```

```python
df_wholesale.clusters.value_counts()
```

```python
df_wholesale[df_wholesale.clusters==2].Region.value_counts()
```

```python
df_wholesale[df_wholesale.clusters==2].describe()
```
